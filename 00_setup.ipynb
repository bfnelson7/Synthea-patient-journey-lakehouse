```sql
-- Catlog: top-level container
CREATE CATALOG IF NOT EXISTS health_lakehouse;

-- Three medallion layers as schemas
CREATE SCHEMA IF NOT EXISTS health_lakehouse.bronze;
CREATE SCHEMA IF NOT EXISTS health_lakehouse.silver;
CREATE SCHEMA IF NOT EXISTS health_lakehouse.gold;

-- Volume in Bronze schema for landing raw CSV files
CREATE VOLUME IF NOT EXISTS health_lakehouse.bronze.raw_synthea;
```
